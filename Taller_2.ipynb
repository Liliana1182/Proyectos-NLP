```python
import os
from zipfile import ZipFile 
```

```python
direccion = 'archivos/poemas.zip'
```

```python
# Descomprimir el archivo ZIP
with ZipFile(direccion) as archivo: 
    archivo.extractall('archivos/') 
```

```python
num_chars = 0
cant_letras_ant = 0

for raiz, dirs, archivos in os.walk('archivos/poemas'):
        for archivo in archivos:
            cant_letras = 0
            with open('archivos/poemas/' + archivo, 'r') as f:
                for line in f:
                    num_chars = len(line)
                    cant_letras = cant_letras + num_chars
                    if cant_letras > cant_letras_ant  :
                        archivo_mayor = cant_letras
                        nombre_archivo = archivo
                        
            cant_letras_ant = cant_letras
            print (archivo, 'Cantidad Palabras',cant_letras)
print ('El archivo con mayor número de palabras es: ',nombre_archivo, 'con',archivo_mayor, 'palabras' )



        
```

# Punto 2:

- Hacer Web Scraping de 10 biografías en Wikipedia (en búcle)
- Obtener el **encabezado** de cada biografía
- Obtener todos los **contenidos** y **etiquetas de título** asociados a los **links** del primer párrafo

```python
import urllib.request
```

<img src='img/carlsagan.png'>

```python
links = ['https://es.wikipedia.org/wiki/Teresa_de_Calcuta','https://es.wikipedia.org/wiki/Rafael_Pombo',
         'https://es.wikipedia.org/wiki/Shakira','https://es.wikipedia.org/wiki/Juanes',
         'https://es.wikipedia.org/wiki/Paulo_Coelho','https://es.wikipedia.org/wiki/Luciano_Pavarotti',
         'https://es.wikipedia.org/wiki/Gilberto_Santa_Rosa','https://es.wikipedia.org/wiki/Jerry_Rivera',
         'https://es.wikipedia.org/wiki/Manuel_Elkin_Patarroyo', 'https://es.wikipedia.org/wiki/Juan_Manuel_Santos']
for link in links:
    request = urllib.request.urlopen(link)
    source = request.read() 
    request.close()
    soup = bs.BeautifulSoup(source,'html.parser') 
    encabezado = soup.find(id='firstHeading')
    print("Encabezado",encabezado.contents)
    for texto in soup.find_all(class_="client-nojs"):
        for parrafo in texto.find_all('p',limit=1):
            for link in parrafo.find_all('a'):
                print (link)
                print("Contenido:")
                print(link.contents)           
                print("Titulo")
                print (link.get('title'))
                print ("  ")
        
```
