# Punto 1:

- `[4pts]` Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con `with open`
- `[4pts]` Convertir a minúsculas
- `[4pts]` Conservar sólo palabras
- `[4pts]` Tokenizar

```python
# Leyendo el archivo
import re
archivo = "UNA_SOLITARIA_VOZ_HUMANA.txt"
with open(archivo, encoding="utf-8") as a:
    texto = a.read()
    #print(texto)
    print (" ")
    print ("TEXTO EN MINUSCULAS")
    texto_mini= texto.lower()  #Convertir a minúsculas    
    print (texto_mini)
    print (" ")
    solo_letras = re.sub(r"[\W\d_]+", " ", texto_mini)
    print ("TEXTO CON SOLO LETRAS")
    print (" ")
    print (solo_letras)
    print (" ")
    print ("TOKENIZACION")
    print (" ")
    texto_token = solo_letras.split()
    print (texto_token)
    
```

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import nltk #Solo se corre una vez
nltk.download('stopwords') #Solo se corre una vez

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
#Quitar palabras vacias
palabras_no_vacias = [palabra for palabra in texto_token if palabra not in stopwords_sp]
palabras_no_vacias


    
```

```python
from collections  import Counter

n = 10
Counter(palabras_no_vacias)
Counter(palabras_no_vacias).most_common(n)
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

```python
from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")
```

```python
texto_stem = []

for palabra in palabras_no_vacias:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
    
texto_stem
```

```python
from collections  import Counter

n = 10
Counter(texto_stem)
Counter(texto_stem).most_common(n)
```
